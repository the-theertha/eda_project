```python
# Plotting all countries with their co2_emissions data

sns.set_theme(style = 'darkgrid')
ax = plt.subplots(figsize=(10,35))
sns.set_color_codes('pastel')

#Plotting graph using seaborn for co2 emission 
sns.barplot(y = co2_emission.index,
            x = co2_emission['2019'])
plt.xlabel('CO2 Emission (metric ton per capita)')
plt.ylabel('Country')
plt.title("CO2 emission by countries in 2019")
```

```python
"""
Getting Top 5 Countries having peak c02 emissions and plotting their history of emissions.
"""

co2_emission['avg'] = co2_emission.mean(axis = 1)
co2_avg = co2_emission.sort_values('avg', ascending = False).head(5)

#Drop 'avg' column 
co2_avg = co2_avg.drop('avg', axis = 'columns')

#Plot time-series linegraph
fig, ax = plt.subplots(figsize =(20, 10))
n = 0
while n <5:
    sns.lineplot( x = co2_avg.columns, y = co2_avg.iloc[n], label = co2_avg.index[n])
    n+=1

plt.xlabel("Year")
plt.ylabel("CO2 Emission (metric ton per capita)")
plt.title("Top 5 Highest Average CO2 Emission per Capita Over Time")
```
